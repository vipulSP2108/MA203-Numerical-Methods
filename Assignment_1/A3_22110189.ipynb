#Problem A3
Newton's law of cooling says that the temperature of a body changes at a rate proportional to the difference between its temperature and that of surrounding medium (the ambient temperature).

dT/dt = −k(T − Ta)

where, T = the temperature of the body (°C), t = time (min), k = the proportionality constant (per minute), and To = the ambient temperature (°C). Suppose that a cup of coffee originally has a temperature of 68 °C. Use Euler's method to compute the temperature from t = 0 to 10 min using step size of 1 min if Ta = 21 °C and k = 0.017 /min

```python
import matplotlib.pyplot as plt

# given values
initial_temp = 68
ambient_temp = 21
k = 0.017
step_size = 1
initial_time = 0
final_time = 10
# step_size = int(input("input step size (que --> 1) "))  # for input from user uncomment below inputs
# initial_time = int(input("input initial time from where we had to find (que --> 0) "))
# final_time = int(input("input final time till where we had to find (que --> 10) "))

# assumed variable
time_val = [initial_time]
temp_val = [initial_temp]
current_time = initial_time
current_temp = initial_temp

# solution using euler
while current_time < final_time:
  change_in_temp = -k * (current_temp - ambient_temp) * step_size
  current_temp = current_temp + change_in_temp
  current_time = current_time + step_size

  time_val.append(current_time)
  temp_val.append(current_temp)


# for printing given values
print("Given,")
print(f"ambient temperature : {ambient_temp}°C")
print(f"step size : {step_size}°C")
print(f"initial temperature : {initial_temp}°C")
print(f"initial time : {initial_time}min")
print(f"final time : {final_time}min")

print("\n")
# for printing table
print("Time (min)   tTemperature (°C)")
print("-----------------------------")
for i in range(len(time_val)):
    time = time_val[i]
    temp = temp_val[i]
    print(f"{time }             {temp:.3f}")

print("\n")
# for plotting results
plt.figure(figsize=(10, 6))
plt.plot(time_val, temp_val, marker="o")
plt.title("Temperature of Coffee Over Time")
plt.xlabel("Time (minutes)")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()
```
