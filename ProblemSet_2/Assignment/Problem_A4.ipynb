#Problem A4
An RLC circuit consists of three elements: a resistor (R), and inductor (L) and a capacitor (C).

fig:

The flow of current across each element induces a voltage drop. Kirchhoff's second voltage law states that the algebraic sum of these voltage drops around a closed circuit is zero:

iR + Ldi/dt +qC= 0

where, i = current, R = resistance, L = inductance, t = time, q = charge and C = capacitance. In addition, the current is related to charge as in

dq/dt = i

a) If the initial values are i(0) = 0 and q(0) = 1 C, use Euler's method to solve this pair of differential equations from t = 0 to 0.1 s using a step size of Δt = 0.01 s. Employ the following parameters for your calculation: R = 200 Ω, L = 5 H and C = 10^4 F. b) Develop a plot of i and q versus t.

```python
import matplotlib.pyplot as plt

# given values
R = 200
L = 5
C = 10**(-4)
initial_i = 0
initial_q = 1
step_size = 0.01
initial_time = 0
final_time = 0.1
# step_size = int(input("input step size (que --> 1) "))  # for input from user uncomment below inputs
# initial_time = int(input("input initial time from where we had to find (que --> 0) "))
# final_time = int(input("input final time till where we had to find (que --> 10) "))


# assumed variable
current_time = initial_time
current_i = initial_i
current_q = initial_q
time_val = [initial_time]
current_val = [initial_i]
charge_val = [initial_q]

# solution using euler
while current_time < final_time:
    change_in_i = (-R * current_i - current_q / C) / L
    change_in_q = current_i
    current_i = current_i + (change_in_i * step_size)
    current_q = current_q + (change_in_q * step_size)
    current_time = current_time + step_size

    time_val.append(current_time)
    current_val.append(current_i)
    charge_val.append(current_q)

# for printing given values
print("Given,")
print(f"resistance : {R}ohm")
print(f"induction : {L}°H")
print(f"capacitor : {C}F")
print(f"initial current : {initial_i}°C")
print(f"initial charge : {initial_q}°C")
print(f"step size : {step_size}sec")
print(f"initial time : {initial_time}min")
print(f"final time : {final_time}min")

print("\n")
# for printing table
print("Time (s)      Current (i)       Charge (q)")
print("----------------------------------")
for i in range(len(time_val)):
    current_time = time_val[i]
    current_i = current_val[i]
    current_q = charge_val[i]
    print(f"{current_time:.3f}            {current_i:.3f}           {current_q:.3f}")

print("\n")
plt.figure(figsize=(10, 6))
plt.plot(time_val, current_val, label='Current (i)')
plt.plot(time_val, charge_val, label='Charge (q)')
plt.title("Current and Charge in RLC Circuit Over Time")
plt.xlabel("Time (seconds)")
plt.ylabel("Current and Charge")
plt.legend()
plt.grid(True)
plt.show()
```
